### 3 . Explain the concept of type 1 and type 2 error in hypothesis testing. provide example scenario of each type 

##### Type-1 Error :
- It is also known as a "False postitive": the error of rejecting a null hypothesis when it is actually true. in other words, this is th error of accepting an alternative hyothesis. when the results can be atrributed to chance. plainly speaking, it occurs when we are observing a difference when in truth there is none. (no statistical significance)

##### Example :
- We conclude that the drug intervention improved symptoms when it actually didn't. these improvements could have arisen from other random factors or measurement error.

##### Type-2 Error : 
- It is also known as "False negative" : the error of not rejecting a null hypothesis when the alternative hypothesis is the true state of nature. in other words, this is the error of failuing to accept an alternative hypothesis when you dont have adequete power. plainly speaking when we are failing to observe a difference when truth is there is one.

##### Example :
- we conclude that the drug intervention did not improve symptoms when it actually did. your study may have missed key indicators of improvements or attributed any improvements to other factors instead.

### 4 . Explain Bayes's Theorem with an example 

##### Bayes Theorem : 
- Bayes thoerem is named after Reverend thomas bayes. it is also known as bayes rule or bayes law. it is used to determine the conditional probability of an event A when event B has already happened. The General statement of bayes theorem is "The conditional probability of an event A, given the occurence of another event B, is equal to the product of event of B,given A and the probability of A divided by the probability of event B"



##### Example : 
- What is the chance of rain during the day ?
the chance of rain given cloud is written as P(rain|cloud)

so using the formula 
##### P(rain|cloud) = P(rain) P(cloud|rain) / P(cloud)
- P(rain) is probability of rain  = 10%
- P(cloud|rain) is probability of cloud, given that rain happens = 50%
- P(cloud) is probability of cloud = 40%

##### P(rain|cloud) = 0.1x0.5 / 0.4 = .125 
or 12.5% chance of rain for the day !

### 7 . Calculate 95% C.I for sample data with mean of 50 and Standard deviation of 5 interpret the results

```python
## importing modules
import scipy.stats as stats

## Given 
sample_mean = 50
standard_deviation = 5
confidence_level = 0.95 #from CI of 95% 

## Assuming sample size of 100
sample_size = 100

## z score
z_score = stats.norm.ppf((1+confidence_level) / 2)

## calculating margin of error 
margin_of_error = z_score * (standard_deviation / (sample_size ** 0.5))

## calculating the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

## printing the results
print(f"95% Confidence Interval : {confidence_interval} ")
print("Interpretation : We are 95% confident that population falls between 49.02 and 50.98 from sample")
```

### 11 . in a survey of 500 people, 65% reported being satisfied with thier current job. calculate 95% confidence interval for the true proportion of people who are satisfied with thier job.

```python
import math

# Given values
sample_proportion = 0.65
sample_size = 500
confidence_level = 0.95

# Z-score for a 95% confidence interval
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_score * math.sqrt((sample_proportion * (1 - sample_proportion))/sample_size)

# Calculate the confidence interval
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)

## results
print(f"95% Confidence Interval: {confidence_interval}")
```
