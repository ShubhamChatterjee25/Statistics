### 2 . write a python function to estimate the population mean using a sample mean and standard deviation. 

```python
## importing requiored modules
import scipy.stats as stats

## Sample mean, standard deviation,Sample size
sample_mean = 50 
sample_std_dev = 10 
sample_size = 100

## we also need Confidence level
confidence_level = 0.95

## Calculate the Z-score for the given confidence level
## we also know that zscore at 95% ci is 1.96
z_score = stats.norm.ppf((1 + confidence_level) / 2)

## Calculate the margin of error
margin_of_error = z_score * (sample_std_dev / (sample_size ** 0.5))

## Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

## Estimated population mean
population_mean = (lower_bound + upper_bound) / 2

## printing the result
print("Estimated Population Mean:", population_mean)
```

### 5 . Write a python script to conduct hypothesis test on the difference between two population means, given a sample from each population

```python
## importing required modules
import numpy as np
from scipy import stats

## Sample data from Population A and Population B
sample_a = np.array([85, 90, 78, 88, 92, 95, 89, 87, 82, 91])
sample_b = np.array([75, 80, 72, 79, 85, 88, 81, 77, 74, 82])

## Set the significance level (alpha)
alpha = 0.05

## Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(sample_a, sample_b)

## Print the test statistic and p-value
print("Test Statistic:", t_stat)
print("P-Value:", p_value)

## Compare the p-value to the significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the population means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the population means.")
```

### 10 . Write python program to calculate the two-sample t-test for independent samples,  given two random samples of equal size and a null hypothesis that the population means are equal 

```python
## Importing the required modules 
from scipy.stats import ttest_ind
import numpy as np

## Creating independent samples of equal size
data1 = np.array([12, 18, 12, 13, 15, 1, 7,  
                20, 21, 25, 19, 31, 21, 17,  
                17, 15, 19, 15, 12, 15])
data2 = np.array([23, 22, 24, 25, 21, 26, 21,  
                21, 25, 30, 24, 21, 23, 19,  
                14, 18, 14, 12, 19, 15])

## Calculating the mean of the two data groups
mean1 = np.mean(data1)
mean2 = np.mean(data2)

## Print mean values
print("Data1 mean value:", mean1)
print("Data2 mean value:", mean2)

## Implementing the t-test
t_test,p_val = ttest_ind(data1, data2)
print("The P-value is: ", p_val)

## hypothesis conclusion using value as 0.05
if p_val < 0.05:      
    print("We can reject the null hypothesis, Population means are not equal")  
else:  
    print("We can accept the null hypothesis, Popualtion means are equal !") 
```

### 13 . A coffee shop owner wants to estimate the daily average revenue for their shop. they take a random sample of 50 days and find the sample mean revenue to be 500$ with a standard devation of 50 dollars. estimate the population mean with 95% confidence interval.  

```python
## Importing required modules
import scipy.stats as stats

## given data
sample_mean = 500
sample_std = 50 # 50 dollars
sample_size = 50 # 50 days
CI = 0.95

## To calculate population mean we need z-score
## Finding Z-score
z_score = stats.norm.ppf((1+CI)/2)

## Getting margin of error using z score
margin_of_error = z_score*(sample_std/sample_size**0.5)

## confidence interval calculation
lower = sample_mean - margin_of_error
upper = sample_mean + margin_of_error

## Estimated population mean is
population_mean = (lower + upper)/2

## printing the results
print("Estimated Population mean for \nSample_mean = 50 and sample_std = 50 is:\n", population_mean)
```
